```python
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
```

```python
def deriv(xi):

    return 2 * xi * np.exp(2*xi)
```

```python
def euler(xi, tau):
    
    return deriv(xi) * tau
```

```python
xi = 0
yi = 1
tau = 0.015
T = 2
N = int(T / tau) + 1

X = np.zeros(N)
y = np.ones(N)
for i in range(1, N):
    yi += euler(xi, tau)
    xi += tau
    
    X[i] = xi
    y[i] = yi
```

```python
fig, ax = plt.subplots(figsize=(10, 8),
                       facecolor='black', subplot_kw={'facecolor' : 'black'})
ax.grid(True, ls='--', alpha=0.6)

c_1 = 3
y_an = c_1 * np.exp(X) + 2*np.exp(2*X)*X - 2*np.exp(2*X)
ax.plot(X, y_an, label='Analytic',
        color='tab:red', lw=4, ls='--', alpha=0.7)
ax.plot(X, y, label='Euler',
        color=cm.magma(0.89), lw=4, alpha=0.7)
ax.set_xlabel('Time', color='white', fontsize=24, fontweight='bold')
ax.set_ylabel('Value', color='white', fontsize=24, fontweight='bold',)
ax.tick_params(axis='x', which='major', colors='white', labelsize=18, rotation=42)
ax.tick_params(axis='y', which='major', colors='white', labelsize=18)

ax.legend(loc='upper left', fontsize=18)

plt.show()
```
